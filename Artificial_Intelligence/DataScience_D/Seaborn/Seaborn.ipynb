```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
## BAR PLOT
```

```python
sns.barplot(x='day',y='tip',data =tips)
```

```python
sns.barplot(x='day',y='total_bill',data =tips)
```

```python
sns.barplot(x='day',y='tip',data =tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',data =tips,hue='sex',palette='winter_r')
```

```python
sns.barplot(x='day',y='tip',data =tips,hue='smoker')
```

```python
sns.barplot(x='day',y='tip',data =tips,palette='spring')
```

```python
sns.barplot(x='day',y='tip',data =tips,palette='spring',order=['Sat','Fri','Thur','Sun'])
```

```python
from numpy import median
sns.barplot(x='day',y='total_bill',data =tips,estimator=median)
```

```python
sns.barplot(x='day',y='tip',data =tips,ci=70)
```

```python
sns.barplot(x='day',y='tip',data =tips,ci=20)
```

```python
sns.barplot(x='day',y='tip',data =tips,capsize=.10,color='green')
```

```python
### DIST PLOT
```

```python
num = np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='green')
```

```python
data = pd.Series(num,name='X')
sns.distplot(data)
```

```python
sns.distplot(data,vertical=True)
```

```python
sns.distplot(data,hist=False)
```

```python
sns.distplot(data,hist=False,rug=True)
```
