# Histogram Plot : Matplotlib (Distplot : Seaborn)

```python
sns.distplot(tips["size"])
plt.show()
```

```python
#bin = 10
sns.distplot(tips["total_bill"], bins=30)
plt.show()
```

```python
#rug = True
sns.distplot(tips["total_bill"], bins=30, rug=True)
plt.show()
```

```python
#hist = False
sns.distplot(tips["total_bill"], bins=30, hist=False)
plt.show()
```

```python
#Kde = False
sns.distplot(tips["total_bill"], bins=30, kde=False)
plt.show()
```

```python
#simple, color, legend
sns.distplot(tips["total_bill"], bins=30, color = 'r', label = "raka")
plt.legend()
plt.show()
```

```python
#fit = norm , color 
sns.distplot(tips["total_bill"], bins=30, fit = norm , color = "r")
plt.show()
```

```python
#vertical = true
sns.distplot(tips["total_bill"], bins=30, fit = norm , color = "r", vertical=True)
plt.show()
```

```python
#norm_hist = true
sns.distplot(tips["total_bill"], bins=30,  norm_hist =True)
plt.show()
```

```python
#xticks = bins, bins = bins

sns.set()

plt.figure(figsize= (16,9))

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.distplot(tips["total_bill"], bins=bins , color = "g")

plt.xticks(bins)
plt.show()
```

```python
#change only bins color : hist_kws (color, linewidth, edgecolor, linestyle="--")


plt.figure(figsize= (16,9))

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.distplot(tips["total_bill"], bins=bins , hist_kws = {"color": "r" , "edgecolor" : "b", "linewidth" : 5}, color = "g")

plt.xticks(bins)
plt.show()
```

```python
#change only bins color : hist_kws (color, linewidth, edgecolor, linestyle="--", alpha, linestyle)


plt.figure(figsize= (16,9))

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.distplot(tips["total_bill"], bins=bins , hist_kws = {"color": "r" , "edgecolor" : "b", "linewidth" : 5
                            , "linestyle": "--" , "alpha" : 0.9}, color = "g")

plt.xticks(bins)
plt.show()
```

```python
#change only bins color : kde_hist (color, linewidth, linestyle="--", alpha, linestyle)


plt.figure(figsize= (16,9))

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.distplot(tips["total_bill"], bins=bins , hist_kws = {"color": "r" , "edgecolor" : "b", "linewidth" : 5
                            , "linestyle": "--" , "alpha" : 0.9}
             ,kde_kws = {"color": "r" , "linewidth" : 5
                            , "linestyle": "--" , "alpha" : 0.9}, color = "g")

plt.xticks(bins)
plt.show()
```

```python
#multiple histogram(displot)

plt.figure(figsize=(12,8))
sns.distplot(tips["total_bill"], bins = 9 , label = "Raka Total bills")
sns.distplot(tips["size"], bins = 9 , label = "Raka size")
sns.distplot(tips["tip"], bins = 9 , label = "Raka Tips")
```

# BarPlot

```python
#per day male and female how much , they pay
sns.barplot(x= tips.day, y = tips.total_bill)
plt.show()
```

```python
#per day male and female how much , they pay, but male paye more with help of hue
sns.barplot(x= tips.day, y = tips.total_bill , hue = tips.sex)
plt.show()
```

```python
#USe directly dataset
sns.barplot(x= "day", y = "total_bill" , hue = "sex" , data = tips)
#plt.xlabel("Size111" , fontsize = 30):  Add the fontsize , also rename the X-axiz / Y-axis 
plt.title("total bill", fontsize = 20)
plt.xlabel("Day" , fontsize = 20)

plt.show()
```

```python
#order : get order a/c to me

order = ["Sun", "Thur", "Fri", "Sat"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot(x= "day", y = "total_bill" , hue = "sex" , data = tips , order = order)

plt.show()
```

```python
#hue order : get hue order a/c to me

order = ["Sun", "Thur", "Fri", "Sat"]

hue_order = ["Female", "Male"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot(x= "day", y = "total_bill" , hue = "sex" , data = tips , order = order, hue_order= hue_order)

plt.show()
```

```python
#estimator :mean , max , etc

order = ["Sun", "Thur", "Fri", "Sat"]

hue_order = ["Female", "Male"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot(x= "day", y = "total_bill" , hue = "sex" , data = tips , order = order, hue_order= hue_order,
           estimator=np.max)

plt.show()
```

```python
#ci = 12 : reduce the size , color

order = ["Sun", "Thur", "Fri", "Sat"]

hue_order = ["Female", "Male"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot(x= "day", y = "total_bill" , hue = "sex" , data = tips , order = order, hue_order= hue_order,
           estimator=np.max , ci = 20 , color = "g")

plt.show()
```

```python
#orientation

order = ["Sun", "Thur", "Fri", "Sat"]

hue_order = ["Female", "Male"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot( x = "total_bill", y = "size" , hue = "sex" ,data = tips ,
            orient = "h")

plt.show()
```

```python
#palatte 

order = ["Sun", "Thur", "Fri", "Sat"]

hue_order = ["Female", "Male"]

plt.title("total bill", fontsize = 10)
plt.xlabel("Day" , fontsize = 10)

sns.barplot( x = "day", y = "total_bill" , hue = "sex" ,data = tips , palette="copper_r")

plt.show()
```

# Scatter : matplotlib (scatterplot : Seaborn)

```python
titanic_df = sns.load_dataset("titanic")
print(titanic_df.info())
print("*"*80)
titanic_df.head(5)
```

```python
sns.scatterplot(x = "age", y = "fare" , data = titanic_df)
plt.show
```

```python
#hue : give more info : male and female

sns.scatterplot(x = "age", y = "fare" , data = titanic_df , hue = "sex")
plt.show()
```

```python
#hue : give more info : male and female
#style : distribute the dataset different style, check easily

sns.scatterplot(x = "age", y = "fare" , data = titanic_df , hue = "sex", style="who")
plt.show()
```

```python
#hue : give more info : male and female
#style : distribute the dataset different style, check easily
#size : give more visualization

plt.figure(figsize=(16,10))

sns.scatterplot(x = "age", y = "fare" , data = titanic_df , hue = "sex", style="who" , size = "who")
plt.show()
```

```python
#hue : give more info : male and female
#style : distribute the dataset different style, check easily
#size : give more visualization

#sizes : increase the sizes of dots (MORE VISIBLE)

plt.figure(figsize=(16,10))

sns.scatterplot(x = "age", y = "fare" , data = titanic_df , hue = "sex", style="who" , size = "who" ,sizes=(100,120))
plt.show()
```

```python
#who live? 80% man : died, more women and chid saved the life
plt.figure(figsize=(16,10))

sns.scatterplot(x = "who", y = "fare" , data = titanic_df , hue = "sex", style="who" , size = "who" ,sizes=(100,120))
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = "age" , y = "fare", data = titanic_df, color = "r")
sns.lineplot(x = "age" , y = "fare", data = titanic_df , color = "g")
sns.barplot(x = "age" , y = "fare", data = titanic_df , color = "y")
plt.show()
```
